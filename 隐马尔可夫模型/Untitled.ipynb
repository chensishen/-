```python
import numpy as np
from hmmlearn import hmm
```

```python
np.random
```

```python
# 生成一些随机数据作为示例
# 假设观测数据为二维连续值
np.random.seed(0)
observations =np.concatenate([                  
    np.random.normal(loc=0.0, scale=0.5, size=(100, 2)),
    np.random.normal(loc=5.0, scale=0.5, size=(100, 2)),
    np.random.normal(loc=10.0, scale=0.5, size=(100, 2))
])
np.random.shuffle(observations)
```

```python
# 创建高斯隐马尔可夫模型
model = hmm.GaussianHMM(n_components=3, covariance_type="full", n_iter=500)
```

```python
# 训练模型
model.fit(observations)
```

```python
# 预测隐状态序列
hidden_states = model.predict(observations)
hidden_states
```

```python
# 输出模型的参数
print("Transition matrix:")
print(model.transmat_)
print("\nMeans and variances of each hidden state:")
for i in range(model.n_components):
    print(f"Hidden state {i}")
    print("Mean = ", model.means_[i])
    print("Variance = ", np.diag(model.covars_[i]))
```

```python
# 假设有3种隐状态(天气)：晴天(0), 多云(1), 雨天(2)
# 有3种观测值(活动)：散步(0), 购物(1), 在家看电影(2)

# 生成一些观察序列作为训练数据
# 观察序列表示一系列活动
observations = np.array([
    [0, 1, 2, 0, 1, 1, 0, 2, 2, 0],   # 第一个观测序列
    [1, 1, 2, 2, 0, 0, 1, 1, 2, 2],   # 第二个观测序列
    [0, 0, 1, 1, 2, 2, 1, 0, 2, 0]    # 第三个观测序列
]).T  # 转置，使其成为适合hmmlearn的输入格式 (每一列是一个时间步)
```

```python
observations = np.array([
    [5, 2, 3],    # 第1个时间步的活动分布
    [4, 3, 3],    # 第2个时间步的活动分布
    [2, 6, 2],    # 第3个时间步的活动分布
    [0, 5, 5],    # 第4个时间步的活动分布
    [3, 3, 4],    # 第5个时间步的活动分布
    [1, 7, 2],    # 第6个时间步的活动分布
    [3, 4, 3],    # 第7个时间步的活动分布
    [5, 3, 2],    # 第8个时间步的活动分布
    [6, 2, 2],    # 第9个时间步的活动分布
    [4, 4, 2],    # 第10个时间步的活动分布
])
```

```python
# 创建多项式隐马尔可夫模型
model = hmm.MultinomialHMM(n_components=3, n_iter=100, init_params="ste")
```

```python
# 初始化模型参数（可选）
model.startprob_ = np.array([0.6, 0.3, 0.1])  # 初始状态概率
model.transmat_ = np.array([
    [0.7, 0.2, 0.1],
    [0.3, 0.5, 0.2],
    [0.3, 0.3, 0.4]
])  # 状态转移矩阵
model.emissionprob_ = np.array([
    [0.6, 0.3, 0.1],
    [0.3, 0.4, 0.3],
    [0.2, 0.2, 0.6]
])  # 观测概率矩阵 (每个状态对应的观测分布)
```

```python
# 训练模型
model.fit(observations)
```

```python
# 测试：对新的观测序列预测最可能的天气状态序列
test_observations = np.array([[3, 3, 4],  # 第1个时间步的活动分布
    [2, 6, 2],  # 第2个时间步的活动分布
    [4, 3, 3], ])
predicted_states = model.predict(test_observations)
```

```python
print("Predicted hidden states (weather):", predicted_states)
```

```python
# 输出模型的参数
print("Transition matrix:")
print(model.transmat_)
print("\nEmission probability matrix:")
print(model.emissionprob_)
```
