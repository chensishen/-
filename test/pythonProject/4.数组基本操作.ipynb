# 生成随机数组np.random

## 正态分布

```python
x1=np.random.normal(1.75,1,100000000)
x1=np.random.randn(100000000)
# x1=np.random.standard_normal()
x1
```

```python
plt.figure(figsize=(20,8),dpi=100)

plt.hist(x1,1000)

plt.show()
```

```python
stock_change=np.random.normal(0,1,[4,5])
```

```python
stock_change
```

## 均匀分布

```python
x2=np.random.uniform(-1,1,100000000)
x2=np.random.rand(100000000)
# x2=np.random.randint(100)
x2
```

```python
plt.figure(figsize=(20,8),dpi=100)

plt.hist(x2,1000)

plt.show()
```
